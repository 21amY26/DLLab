Implementing AND gate using a simple feed forward neural network.
Sigmoid function will be used since it gives output between 0 and 1 -> better for prob. based classfn.

```python
import numpy as np
```

```python
class SimpleNeuralNw:
    def __init__(self, ip_size, alpha=0.1): #alpha is learning rate
        self.weights = np.random.randn(ip_size)
        self.bias = 0
        self.alpha = alpha

    def sigmoid(self,x):
        return 1/(1 + np.exp(-x))

    def sigmoid_derivative(self,x):
        return x*(1-x)
    
    def predict(self, ips):
        # sum = weighted sum + bias_term
        sum = np.dot(ips, self.weights) + self.bias
        return self.sigmoid(sum)

    def train(self, train_ips, labels, epochs):
        for epoch in range(epochs):
            for ips, label in zip(train_ips,labels):

                fwd_pred = self.predict(ips)

                error = label - fwd_pred

                adjustment = error * self.sigmoid_derivative(fwd_pred)

                self.weights += self.alpha * adjustment * ips
                self.bias += self.alpha*adjustment
```

```python
# training nw on AND gate
input = np.array([[0,0],[0,1],[1,0],[1,1]])
labels = np.array([0,0,0,1])

perceptron = SimpleNeuralNw(ip_size=2)
perceptron.train(input, labels, epochs=10000)
```

```python
#testing
for i in input:
    print(f"Input: {i}\nPrediction: {perceptron.predict(i):.3f}")
```
